### NLTK

```python
def get_sentiment(row, k):
    sid = SentimentIntensityAnalyzer()
    sentiment_score = sid.polarity_scores(row)

    positive_meter = round((sentiment_score['pos'] * 10), 2)
    negative_meter = round((sentiment_score['neg'] * 10), 2)

    return positive_meter if k == 'positive' else negative_meter
```

```python
df['positive'] = df["sentence"].apply(get_sentiment, k='positive')
df['negative'] = df["sentence"].apply(get_sentiment, k='negative')
```

### TextBlob

```python
def get_sentiment2(row):
    return TextBlob(row).sentiment.polarity
```

```python
df["polarity"] = df["sentence"].apply(get_sentiment2)

spoilers2 = df[df["has_spoiler"] == 1]
safe2 = df[df["has_spoiler"] == 0]
```

```python
sppa = spoilers2["polarity"].mean()

sppm = float(spoilers2["polarity"].mode())

sfpa = safe2["polarity"].mean()

sfpm = float(safe2["polarity"].mode())

print(f"Means:\nSpoiler: {sppa}\nSafe: {sfpa}\n\nModes:\nSpoiler: {sppm}\nSafe: {sfpm}")
```
